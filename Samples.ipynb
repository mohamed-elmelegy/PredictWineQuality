# Import packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Reading Data from csv to Pandas

```python
#Reading csv data as Dataframe

Wine_Quality=pd.read_csv("Wine_Quality.csv")
```

# Generate random samples and get their Means and Variances

Generate 50 different random sample of 50 observations and get Sample Mean and Sample Variance for each one of them 

```python
#Getting samples and calculate their means and variances

observations = {}
for i in range(50):
    sample = Wine_Quality.sample(50)
    smean = sample.mean()
    svar = sample.var()
    observations[i] = {'mean':smean, 'var':svar}
    #print('Sample Mean #'+str(i+1)+ (':\n****************\n') +str( smean), "\n***********************************")
    #print('Sample Variance #'+str(i+1)+ (':\n****************\n') +str( svar), "\n***********************************")
    
#print(observations)
```

# Scatter Plot for 50 Sample Means with each feature

```python
#Plotting scatter for sample Mean vs samples number for each random variable

for col in Wine_Quality:
    for i in range(50):
        plt.scatter(observations[i]['mean'][col],i)
        plt.title(col)
    plt.show()
```

# Scatter Plot for 50 Sample Variances with each feature

```python
#Plotting scatter for sample Variance vs samples number for each random variable

for col in Wine_Quality:
    for i in range(50):
        plt.scatter(observations[i]['var'][col],i)
        plt.title(col)
    plt.show()
```

# Plot Mean of the 50 Sample Means

```python
#Plot mean of samples means vs Real mean for each rv

#for Mean of samples means
mean_of_means = []
print("Mean of samples means:\n************************\n")
for col in Wine_Quality:
    tmp_mean = []
    for i in range(50):
        tmp_mean.append(observations[i]['mean'][col])
    plt.subplots(figsize=(10,5))
    mean_of_means.append([col, np.mean(tmp_mean)])
    sns.distplot(tmp_mean, kde=True, hist=True, rug=True, bins=15, color='r')
    plt.title(col+' => Mean of Samples Mean')
    plt.show()

tmp = pd.DataFrame(mean_of_means)
tmp = tmp.rename(columns={0:'column', 1:'mean'})
print(tmp)
tmp.plot(kind='hist', bins=20)
```
