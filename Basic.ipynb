# Importing some packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Reading csv data as Dataframe 

```python
Wine_Quality=pd.read_csv("Wine_Quality.csv")
#Wine_Quality.describe()
```

# Print the real Mean and Variance of the whole data

```python
# 1)Mean && 2) Variace for the whole dataset  
Mean = Wine_Quality.mean()
Variance = Wine_Quality.var()
print("****************************************************************")
print("Mean for the whole data:\n************************\n", Mean)
print("****************************************************************")
print("Variance for the whole data:\n****************************\n", Variance)
print("****************************************************************")
```

# Scatter plot between each pair of data columns

Scatter plot for distinguish the correlation between each pair of features

```python
column_lable={0:'fixed acidity',1:'volatile acidity',2:'citric acid',
              3:'residual sugar',4:'chlorides',5:'free sulfur dioxide',
              6:'total sulfur dioxide',7:"density" ,8:'pH',9:"sulphates",
              10:'alcohol',11:'quality'}


# Scatter between each pair of columns
for i in range(12):
    for j in range(i+1, 12):
        plt.scatter(x=column_lable[i], y=column_lable[j], data=Wine_Quality, c='y', edgecolors='black')
        plt.title(column_lable[i]+' & '+column_lable[j])
        plt.show()
```

# Histogram and Rug Plot for each feature to understand its distribution 

```python
# 2) Histogram && 4)Rug Plot
for i in range(12):
   sns.distplot(Wine_Quality[column_lable[i]],rug=True,bins=20,kde=True,hist=True)
   plt.title(column_lable[i])
   plt.show()
```

# Box Plot for each feature

to get the hot values or quarters of the distribution

```python
#3)Boxplot
for i in range(12):
   plt.boxplot(Wine_Quality[column_lable[i]], widths = 0.95, showmeans =True)
   plt.title(column_lable[i])
   plt.show()

Wine_Quality.plot(kind='box', title='Box Plot', sharex=False, sharey=False, figsize=(18,5))
```

# Count Plot for unique labels in data

```python
print("Unique values in lable column: ", Wine_Quality['quality'].unique())
sns.countplot(x='quality', data=Wine_Quality)
```

# Plot features tedency vs label

```python
for i, col in enumerate(Wine_Quality.columns):
    if(col == 'quality'):
        continue
    plt.figure(i)
    sns.barplot(x='quality', y =col, data=Wine_Quality)
```

# Print Covariance

```python
#covariance Matrix

Cov = Wine_Quality.cov()
print("****************************************************************")
print("Cov for all pairs of columns:\n************************\n", Cov)
```

# Plot Correlation Matrix

```python
plt.subplots(figsize=(10,10))
sns.heatmap(Wine_Quality.corr(), mask=np.zeros_like(Wine_Quality.corr(), dtype=np.bool), cmap=sns.diverging_palette(220, 10, as_cmap=True),
            square=True, annot=True)
plt.show()
```
